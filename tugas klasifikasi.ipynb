IMPORT LIBRARY

```python
import numpy as np
import pandas as pd
```

IMPORT DATASET

```python
dataset = pd.read_csv('StudentPerformance.csv')
dataset
```

MEMASUKKAN FITUR DAN LABEL

```python
x = dataset.drop('GradeClass', axis = 1)
y = dataset['GradeClass']
```

MEMBUAT DATA TRAINING DAN TESTING

```python
from sklearn.model_selection import train_test_split, GridSearchCV

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

MENSTANDARISASI DATA 

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.fit_transform(x_test)
```

TUNING DATA MELALUI GRID SEARCH

```python
param_grid = {
    'C' : [0.1, 1, 10, 100],
    'gamma' : [1, 0.1, 0.01, 0.001],
    'kernel' : ['linear', 'rbf']
}
```

MELATIH MODEL SVM

```python
from sklearn.svm import SVC

grid_Search = GridSearchCV(SVC(class_weight='balanced'), param_grid, refit=True, verbose=2)
grid_Search.fit(x_train_scaled, y_train)

model_svm = grid_Search.best_estimator_
y_pred= model_svm.predict(x_test_scaled)
```

MELATIH MODEL DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier

model2 = DecisionTreeClassifier()
model2.fit(x_train_scaled, y_train)

y_pred2 = model2.predict(x_test_scaled)
```

MELATIH MODEL DECISION NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB

model3 = GaussianNB()
model3.fit(x_train_scaled, y_train)

y_pred3 = model3.predict(x_test_scaled)
```

MELAKUKAN METRIK EVALUASI TERHADAP 3 MODEL

```python
from sklearn.metrics import accuracy_score, classification_report

SVM_accuracy = accuracy_score(y_test, y_pred)
DecisionTree_accuracy = accuracy_score(y_test, y_pred2)
NaiveBayes_accuracy = accuracy_score(y_test, y_pred3)

print('Akurasi SVM: ', SVM_accuracy)
print('Akurasi Decision tree: ', DecisionTree_accuracy)
print('Akurasi Naive Bayes: ', NaiveBayes_accuracy)
```

```python
print('SVM:', classification_report(y_test, y_pred))
print('Decision Tree:', classification_report(y_test, y_pred2))
print('Naive Bayes:', classification_report(y_test, y_pred3))
```
